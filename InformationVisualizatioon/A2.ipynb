```python
country_df = df.groupby(['country', 'gender']).size().reset_index(name='count')
country_df
```

```python
gender_df = pd.merge(left=country_df[country_df['gender']=="M"],right=country_df[country_df['gender']=="F"],how='inner',on='country')
```

```python
gender_df = gender_df.rename(columns={'gender_x': 'male', 'count_x':'count_male', 'gender_y': 'female', 'count_y': 'count_female'})
gender_df['ratio'] = gender_df['count_male'] / gender_df['count_female']
gender_df['total'] = gender_df['count_male'] + gender_df['count_female']
pd.DataFrame(gender_df, columns=['country', 'total', 'count_female', 'count_male', 'ratio'])
```

```python
sort_gender_df = gender_df.sort_values(by=['total'])

# sort_gender_df.plot(x='country', y='total', kind='bar')
sort_gender_df.loc[sort_gender_df['ratio'] < 1, 'ratio_type'] = 'low'
sort_gender_df.loc[sort_gender_df['ratio'] >= 2, 'ratio_type'] = 'high'
sort_gender_df.loc[(sort_gender_df['ratio'] >=1) & (sort_gender_df['ratio'] < 2), 'ratio_type'] = 'medium'

pd.DataFrame(sort_gender_df, columns=['country', 'total', 'count_female', 'count_male', 'ratio_type'])
```

```python
# sort_gender_df.plot(x='country', y='ratio', kind='bar')

sns.set(style="white",context="talk")
fig,(ax1,ax2) = plt.subplots(2,1,figsize=(10,6),sharex=True)

sns.barplot(x=sort_gender_df['country'], y=sort_gender_df['total'], palette="deep", ax=ax1)  # palette :rocket husl deep  vlag
 
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Total positions")
# ax1.spines['bottom'].set_visible(False)
 

sns.barplot(x=sort_gender_df['country'], y=sort_gender_df['ratio'], palette="deep", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("Male to female ratio")
# ax2.spines['bottom'].set_visible(False)

sns.despine(bottom=True)
# plt.setp(fig.axes, yticks=[])
plt.suptitle('The distribution of positions per country')
plt.tight_layout(h_pad=2)
plt.xticks(rotation=-90)
```

```python

# sort_gender_df.plot.scatter(x='count_female', y='count_male', c='country', colormap='jet')
sns.scatterplot(data=sort_gender_df, x='total', y='ratio', hue='country', style='ratio_type', palette="deep")
plt.legend(bbox_to_anchor=(2.01, 1),borderaxespad=0)
plt.title('Gender distribution of positions per country')
sns.despine()
```
